```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
import pandas as pd
df = pd.read_csv("data/customer_churn_data.csv")  # Charger un fichier CSV
```

```python
print(df.columns)
```

```python
df=df.drop(columns=['customerID','gender', 'SeniorCitizen', 'Partner', 'Dependents','PhoneService', 'MultipleLines', 'InternetService',
       'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
       'StreamingTV', 'StreamingMovies','PaperlessBilling',
       'PaymentMethod','TotalCharges'])
```

```python
print(df.columns)
```

```python
df.isna().sum()
```

```python
df['Churn'] = df['Churn'].replace({'Yes': 1, 'No': 0})
```

```python
print(df.head())
```

```python
df['Contract'] = df['Contract'].replace({'Month-to-month': 0, 'One year': 1,'Two year':2})
```

```python
X=df[["tenure","Contract","MonthlyCharges"]]
y=df[["Churn"]]
```

```python
model=LinearRegression()
```

```python
model.fit(X,y)
```

```python
new_data = pd.DataFrame({'tenure': [0], 'Contract': [0], 'MonthlyCharges': [0]})
prediction = model.predict(new_data)

print("Churn :", prediction[0])
```

```python
joblib.dump(model, "model/regression_model.pkl")
```
